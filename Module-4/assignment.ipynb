#Q Explain Key Statistical Concepts

1. Population and Sample

Population: The complete set of all items or people you want to study.
Example: All customers of a car rental company.

Sample: A smaller subset taken from the population.
Example: Rental durations of 50 customers.

Why sample?
Because studying the entire population is difficult, costly, or time-consuming.


2. Variable

A variable is any characteristic or value that can change.

Examples:

Age

Height

Rental duration

Salary

Types:

Numerical variables → numbers (age, marks)

Categorical variables → categories (gender, color)


3. Mean (Average)

The mean is the average value.

Formula:

Mean = Sum of all values / Number of values

Use: Shows the central value of data.


4. Median

The median is the middle value when data is arranged in order.

Example:

Data: 1, 3, 5 → Median = 3

If even numbers:

Data: 1, 3, 5, 7
Median = (3+5)/2 = 4

Use: Better when data has outliers.


5. Mode

The mode is the value that appears most frequently.

Example:

Data: 2, 3, 3, 5, 6
Mode = 3

Use: Shows most common value.


6. Range

The range shows the spread of data.

Formula:

Range=Maximum−Minimum

Example:

Data: 2, 5, 9

Range = 9 − 2 = 7


7. Variance

Variance measures how far values are spread from the mean.

Formula:


Variance = ∑(x−mean)^2 ? n
	​
Higher variance → more spread
Lower variance → less spread


8. Standard Deviation

Standard deviation is the square root of variance.

Standard deviation= sqrt(Variance)
	​
Use: Most important measure of data spread.

Low SD → data close to mean
High SD → data widely spread


9. Probability

Probability measures the chance of an event.

Formula:

Probability = Favorable outcomes / Total outcomes


10. Distribution

Distribution shows how data values are spread.

Common type:
Normal Distribution (Bell curve)

Properties:

Mean = Median = Mode

Symmetrical shape

Used in ML, statistics, and real-world analysis.


11. Outliers

Outliers are unusual or extreme values.

Example:

Data: 2, 3, 4, 5, 100

Here, 100 is an outlier.

Outliers affect mean and analysis.


12. Frequency

Frequency = how many times a value occurs.

Example:

Data: 2, 2, 3, 4

Frequency of 2 = 2

```python
# Q1  Business Problem: A car rental company wants to analyze the rental durations of itscustomers to understand the typical rental period and optimize its pricing andfleet 
# management strategies. Data: Let's consider the rental durations (in days) for asample of 50 customers:
# 3, 2, 5, 4, 7, 2, 3, 3, 1, 6, 4, 2, 3, 5, 2, 4, 2, 1, 3, 5, 6, 3, 2, 1, 4, 2, 4, 5, 3, 2, 7, 2, 3, 4, 5, 1, 6, 2, 4, 3, 5, 3, 2, 4, 2, 6, 3, 2, 4, 5

# a) Mean: What is the average rental duration for customers at the car rental company?
# b) Median: What is the typical or central rental duration experiencedbycustomers?
# c) Mode: Are there any recurring or most frequently occurring rental durations for customers?


rental = [3,2,5,4,7,2,3,3,1,6,4,2,3,5,2,4,2,1,3,5,6,3,2,1,4,2,4,5,3,2,7,2,3,4,5,1,6,2,4,3,5,3,2,4,2,6,3,2,4,5]

mean = np.mean(rental)
median = np.median(rental)
mode = stats.mode(rental, keepdims=True)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode.mode[0])

```

```python
# Q2 An e-commerce platform wants to analyze the delivery times of its shipments tounderstand the variability in order fulfillment and optimize its logistics operations. Data:
# Let's consider the delivery times (in days) for a sample of 50 shipments:
# 3, 5, 2, 4, 6, 2, 3, 4, 2, 5, 7, 2, 3, 4, 2, 4, 2, 3, 5, 6, 3, 2, 1, 4, 2, 4, 5, 3, 2, 7, 2, 3, 4, 5, 1, 6, 2, 4, 3, 5, 3, 2, 4, 2, 6, 3, 2, 4, 5, 3

# Questions:
# a. Range: What is the range of the delivery times?
# b. Variance: What is the variance of the delivery times?
# c. Standard Deviation: What is the standard deviation of the deliverytimes?

delivery = [3,5,2,4,6,2,3,4,2,5,7,2,3,4,2,4,2,3,5,6,3,2,1,4,2,4,5,3,2,7,2,3,4,5,1,6,2,4,3,5,3,2,4,2,6,3,2,4,5,3]

range_val = np.max(delivery) - np.min(delivery)
variance = np.var(delivery)
std_dev = np.std(delivery)

print("Range:", range_val)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Q10 The lifetimes of a certain brand of light bulbs are normally distributed witha meanof 1000 hours and a standard deviation of 100 hours. 
# What is the probabilitythat arandomly selected light bulb lasts between 900 and 1100 hours? 
# Data: Meanlifetime(μ) = 1000 hours, Standard deviation (σ) = 100 hours, Lifetime range (lower limit x1, upper limit x2)

mean = 1000
std = 100

prob = norm.cdf(1100, mean, std) - norm.cdf(900, mean, std)

print("Probability:", prob)

```

```python
# Q11. A study is conducted to estimate the mean height of a population. Arandomsampleof 100 individuals is selected, and their heights are measured. 
# Calculate a 95%confidence interval for the population mean height, given that the sample meanheight is 170 cm and the sample standard deviation is 8 cm. 
# Data: Sample size(n) =100, Sample mean (x̄) = 170 cm, Sample standard deviation (s) = 8 cm, Confidencelevel = 95%


mean = 170
std = 8
n = 100

z = 1.96

margin = z * (std/np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("Confidence Interval:", lower, "to", upper)
```

```python
# Q12. A manufacturing company claims that the average weight of its product is 500grams. To test this claim, a random sample of 25 products is selected, and their
# weights are measured. The sample mean weight is found to be 510 grams withasample standard deviation of 20 grams. Perform a hypothesis test to determineif
# there is evidence to support the company's claim. 
# Data: Sample size (n) =25, Sample mean (x̄) = 510 grams, Sample standard deviation (s) = 20 grams, Population mean (μ) = 500 grams


sample_mean = 510
pop_mean = 500
std = 20
n = 25

z = (sample_mean - pop_mean)/(std/np.sqrt(n))

print("Z-score:", z)

```

```python
# Q13. A researcher wants to test whether a new teaching method improves student performance. 
# A random sample of 50 students is divided into two groups: onegrouptaught using the new method and the other using the traditional method. The averagetest scores of the two groups are compared. State the null and alternative hypothesesfor this study. 
# Data: Sample size (n) = 50, Test scores of the two groups


group_new = np.array([85, 88, 79, 92, 85, 90, 84, 87, 83, 89, 
                      91, 78, 86, 88, 82, 85, 87, 90, 84, 86, 
                      89, 81, 85, 87, 88])

group_traditional = np.array([78, 82, 75, 80, 83, 79, 81, 77, 85, 76, 
                            82, 80, 79, 81, 78, 84, 77, 80, 79, 82, 
                            75, 81, 80, 78, 76])

# 2. Perform the Independent Samples t-test
# equal_var=True assumes the populations have the same variance
t_stat, p_value = stats.ttest_ind(group_new, group_traditional)

# 3. Output the results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# 4. Interpret the results (using alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the Null Hypothesis (Significant improvement).")
else:
    print("Conclusion: Fail to reject the Null Hypothesis (No significant difference).")
```

```python
# 14. A survey is conducted to estimate the proportion of people in a city who support aparticular policy. 
# A random sample of 500 individuals is surveyed, and 320of themexpress support for the policy. Calculate a 90% confidence interval for the
# population proportion, given the sample proportion. 
# Data: Sample size (n) =500, Number of successes (x) = 320, Confidence level = 90%


n = 500
x = 320

p = x/n
z = 1.645

margin = z*np.sqrt((p*(1-p))/n)

lower = p - margin
upper = p + margin

print("Confidence Interval:", lower, "to", upper)
```

```python
# Q15 A company sells smartphones, and the number of defects per batch follows a
# Poisson distribution with a mean of 2 defects. What is the probability of havingexactly 3 defects in a randomly selected batch? 
# Data: Mean number of defects (λ) =2, Number of defects (x) = 3



lam = 2
x = 3

prob = poisson.pmf(x, lam)

print("Probability:", prob)
```
